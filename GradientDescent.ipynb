**Gradient Descent from Scratch**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Generate a Simple Dataset (y=2x+1 - No Noise)**

```python
np.random.seed(42)
X = np.linspace(0, 10, 50) # 50 Evenly Spaced Points Between 0 to 10)
true_m = 2
true_c = 1
y = true_m * X + true_c
```

**Plotting Input Data**

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Points", color="blue")
plt.title("Input Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

**Initializing the Parameters**

```python
m = 0.0
c = 0.0
learning_rate = 0.01 # Learning Rate
iterations = 2001 # Number of Iterations
```

**Store Loss for Plotting**

```python
loss_history = []
```

**Gradient Descent**

```python
for i in range(iterations):
  # Predicted y values
  y_pred = m * X + c

  # Compute Gradients
  m_gradient = (-2 / len(X)) * np.sum(X * (y - y_pred))
  c_gradient = (-2 / len(X)) * np.sum(y - y_pred)

   # Update Parameters
  m = m - learning_rate * m_gradient
  c = c - learning_rate * c_gradient

  # Compute Loss [MSE]
  loss = np.mean((y - y_pred) ** 2)
  loss_history.append(loss)

  # Print Updates Every 100 Iterations
  if i % 100 == 0:
    print(f"Iteration {i}: Loss= {loss:}|| m= {m:}|| c= {c:}")
```

**Plotting**

```python
# Print Final Parameters
print(f"Final Parameters: m= {m:.4f}|| c= {c:.4f}")

# Plot the Dataset and the Fitted Line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Points", color="blue")
plt.plot(X, m * X + c,label=f"Fitted Line: y = {m:.4f}x + {c:.4f}", color="red")
plt.title("Gradient Descent: Line Fitting")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Loss History

plt.figure(figsize=(10, 6))
plt.plot(range(iterations), loss_history, label="Loss", color="red")
plt.title("Loss History")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.legend()
```

**Generate a Simple Dataset (y=2x+1 with Noise)**

```python
np.random.seed(42)
X = np.linspace(0, 10, 50) # 50 Evenly Spaced Points Between 0 to 10)
true_m = 2
true_c = 1
y = true_m * X + true_c + np.random.normal(0, 2, X.shape) # Add Some Noise
```

**Plotting Input Data**

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Points", color="blue")
plt.title("Input Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

**Initialize Parameters (Slope & Intercept)**

```python
m = 0.0
c = 0.0
learning_rate = 0.001 # Learning Rate
iterations = 2001 # Number of Iterations
```

**Store Loss for Plotting**

```python
loss_history = []
```

**Gradient Descent**

```python
for i in range(iterations):
  # Predicted y values
  y_pred = m * X + c

  # Compute Gradients
  m_gradient = (-2 / len(X)) * np.sum(X * (y - y_pred))
  c_gradient = (-2 / len(X)) * np.sum(y - y_pred)

   # Update Parameters
  m = m - learning_rate * m_gradient
  c = c - learning_rate * c_gradient

  # Compute Loss [MSE]
  loss = np.mean((y - y_pred) ** 2)
  loss_history.append(loss)

  # Print Updates Every 100 Iterations
  if i % 100 == 0:
    print(f"Iteration {i}: Loss= {loss:}|| m= {m:}|| c= {c:}")
```

**Plotting**

```python
# Print Final Parameters
print(f"Final Parameters: m= {m:.4f}|| c= {c:.4f}")

# Plot the Dataset and the Fitted Line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Data Points", color="blue")
plt.plot(X, m * X + c,label=f"Fitted Line: y = {m:.4f}x + {c:.4f}", color="red")
plt.title("Gradient Descent: Line Fitting")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Loss History

plt.figure(figsize=(10, 6))
plt.plot(range(iterations), loss_history, label="Loss", color="red")
plt.title("Loss History")
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.legend()
```
